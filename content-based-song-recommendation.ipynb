```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
#importing libraries
```

```python

# making data frame from csv file
songs_df = pd.read_csv("/kaggle/input/spotify-dataset/cleaned_dataset.csv")
  
# sorting by first name
songs_df.sort_values("Track", inplace=True)
  
# dropping ALL duplicate values
songs_df.drop_duplicates(subset="Track",
                     keep= 'first', inplace=True)
  
# displaying data
songs_df
```

```python
k = 11 #number of songs to recommend (n + 1)
knn = NearestNeighbors(n_neighbors = k) #machine learning model ultilised is KNN
knn.fit(songs_df[["Energy","Loudness","Speechiness","Acousticness","Instrumentalness","Valence","Tempo"]])
#number of features on which songs are recommended
```

```python
def get_song_rec(song_name, knn, songs_df):
#     songs_df = pd.read_csv("/kaggle/input/spotify-dataset/cleaned_dataset.csv")
    
    vector = songs_df[["Energy","Loudness","Speechiness","Acousticness","Instrumentalness","Valence","Tempo"]].loc[songs_df["Track"] == song_name].values.reshape(1,-1)
    
    neighbors = knn.kneighbors(vector, return_distance = False)
#     print(neighbors)
    similar_songs = songs_df.iloc[neighbors[0]].reset_index(drop= True)
#     similar_songs = songs_df.iloc[1:]
#     print(similar_songs)
    return similar_songs
#songs are recommended on the basis of their vector values and their closeness to the input songs.
```

```python
if __name__ == "__main__":
    song_name = input("Enter a song name: ")
    similar_songs = get_song_rec(song_name, knn, songs_df)
    # print("The most similar songs to {} are: {}".format(song_name, similar_songs))
    print(similar_songs['Track'].tail(-1))
    
```
